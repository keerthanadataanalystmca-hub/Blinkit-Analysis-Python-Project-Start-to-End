### BUSINESS REQUIREMENTS

#### KPI's REQUIREMENTS

```python
# Assuming 'df' is your pandas DataFrame and it contains 'Sales' and 'Rating' columns.

# Calculate Total Sales
total_sales = df['Sales'].sum()

# Calculate Average Sales
avg_sales = df['Sales'].mean()

# Calculate the number of items sold (count of non-null entries in 'Sales')
no_of_items_sold = df['Sales'].count()

# Calculate Average Ratings
avg_ratings = df['Rating'].mean()

# Display Total Sales formatted as currency with commas and two decimal places
print(f"Total Sales: ${total_sales:,.0f}")
print(f"Average Sales: ${avg_sales:,.2f}")
print(f"Number of Items Sold: {no_of_items_sold}")
print(f"Average Ratings: {avg_ratings:.2f}")


```

#### CHARTS REQUIREMENTS

#### TOTAL SALES BY FAT CONTENT

```python
# Group the data by 'Item Fat Content' and calculate the sum of 'Sales' for each group.
sales_by_fat = df.groupby('Item Fat Content')['Sales'].sum()

# Create a pie chart using the grouped data.
plt.pie(sales_by_fat, labels=sales_by_fat.index, autopct='%.1f%%', startangle=90)

# Add a title to the pie chart.
plt.title('Sales by Fat Content')

# Ensure the pie chart is a circle.
plt.axis('equal')

# Display the plot.
plt.show()
```

#### TOTAL SALES BY FAT CONTENT

```python

# Group the DataFrame by 'Item Type', sum the 'Sales' for each type, and sort the results in descending order.
sales_by_type = df.groupby('Item Type')['Sales'].sum().sort_values(ascending=False)

# Create a new figure for the plot with a specified size (10x6 inches).
plt.figure(figsize=(10, 6))

# Create a bar chart with 'Item Type' on the x-axis and 'Total Sales' on the y-axis.
bars = plt.bar(sales_by_type.index, sales_by_type.values)

# Rotate the x-axis labels by 90 degrees to prevent overlapping.
plt.xticks(rotation=-90)

# Set the x-axis and y-axis labels.
plt.xlabel('Item Type')
plt.ylabel('Total Sales')

# Set the title of the plot.
plt.title('Total Sales by Item Type')

# Loop through each bar in the bar chart to add a text label for its height.
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, 
             f'{yval:,.0f}', 
             ha='center', va='bottom', fontsize=8)

# Adjust the plot layout to make sure all elements (like labels) fit without being cut off.
plt.tight_layout()

# Display the plot.
plt.show()
```

#### OUTLET TIER BY ITEM FAT CONTENT

```python

# Group the data by 'Outlet Location Type' and 'Item Fat Content', sum the 'Sales', and then reshape the data using unstack().
# This creates a DataFrame where 'Outlet Location Type' is the index, and 'Item Fat Content' values ('Regular' and 'Low Fat') are columns.
grouped = df.groupby(['Outlet Location Type', 'Item Fat Content'])['Sales'].sum().unstack()

# Reorder the columns to place 'Regular' first, followed by 'Low Fat'.
grouped = grouped[['Regular', 'Low Fat']]

# Create a bar chart from the grouped DataFrame.
# The `kind='bar'` argument tells pandas to plot a bar chart.
ax = grouped.plot(kind='bar', figsize=(8, 5), title='Outlet Tier by Item Fat Content')

# Set the x-axis label.
plt.xlabel('Outlet Location Tier')

# Set the y-axis label.
plt.ylabel('Total Sales')

# Set the title for the legend.
plt.legend(title='Item Fat Content')

# Adjust the plot layout to make sure all elements (like labels and legend) fit nicely.
plt.tight_layout()

# Display the plot.
plt.show()
```

#### TOTAL SALE BY OUTLET ESHTABLISHMENT

```python
# Group data by 'Outlet Establishment Year', sum the 'Sales', and sort by the year.
sales_by_year = df.groupby('Outlet Establishment Year')['Sales'].sum().sort_index()

# Create a new figure with a specified size.
plt.figure(figsize=(9, 5))

# Create a line plot with markers ('o') and a dashed linestyle ('--').
plt.plot(sales_by_year.index, sales_by_year.values, marker='o', linestyle='--')

# Set the x and y axis labels and the plot title.
plt.xlabel('Outlet Establishment Year')
plt.ylabel('Total Sales')
plt.title('Outlet Establishment')

# Add text labels to each data point on the line plot.
for x, y in zip(sales_by_year.index, sales_by_year.values):
    plt.text(x, y, f'{y:,.0f}', ha='center', va='bottom', fontsize=8)

# Adjust the plot layout to prevent labels from being cut off.
plt.tight_layout()

# Display the first plot.
plt.show()
```

#### SALES BY OUTLET SIZE

```python
# Group data by 'Outlet Size' and sum the 'Sales'.
sales_by_size = df.groupby('Outlet Size')['Sales'].sum()

# Create a new figure with a specified size.
plt.figure(figsize=(4, 4))

# Create a pie chart with labels, autopct for percentages, and a start angle.
plt.pie(sales_by_size, labels=sales_by_size.index, autopct='%.1f%%', startangle=90)

# Set the title of the pie chart.
plt.title('Outlet Size')

# Adjust the plot layout.
plt.tight_layout()

# Display the second plot.
plt.show()
```

#### SALES BY OUTLET LOCATION

```python

# Group the DataFrame by 'Outlet Location Type', sum the 'Sales', and reset the index to turn the grouped result into a new DataFrame.
sales_by_location = df.groupby('Outlet Location Type')['Sales'].sum().reset_index()

# Sort the new DataFrame by 'Sales' in descending order to prepare for the bar plot.
sales_by_location = sales_by_location.sort_values('Sales', ascending=False)

# Create a new figure with a specified size.
plt.figure(figsize=(8, 3))

# Create a bar plot using Seaborn, with 'Sales' on the x-axis and 'Outlet Location Type' on the y-axis.
ax = sns.barplot(x='Sales', y='Outlet Location Type', data=sales_by_location)

# Set the title and labels for the plot. Note the reversal of x and y labels for a horizontal bar chart.
plt.title('Total Sales by Outlet Location Type')
plt.xlabel('Total Sales')
plt.ylabel('Outlet Location Type')

# Adjust the plot layout to ensure everything fits.
plt.tight_layout()

# Display the plot.
plt.show()
```
